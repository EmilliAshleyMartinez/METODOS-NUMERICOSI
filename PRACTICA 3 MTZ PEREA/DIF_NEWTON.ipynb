```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
#MARTINEZ PEREA EMILLI
```

```python
#def funcion(x):
#DATOS EJEMPLO 1 P.94
x=np.array([1.0 ,1.3 ,1.6 ,1.9 ,2.2])
    #return np.exp(-x)-x

#def funciong(x):
y=np.array([0.7651977 ,0.6200860 ,0.4554022 ,0.2818186 ,0.1103623])
  #return np.exp(-x)

n=x.size
p=np.zeros((n,n))
p[:,0]=y
w=0
```

```python
print("# i\t xi \t f(xi) \t  1RAS DIF \t 2DAS DIF \t 3RAS DIF \t 4TAS DIF \t 5TAS DIF" )
#Tabla de diferencias divididas
for w in range (n):
  for i in range (1,n):
    for j in range (0,n-i):
      p[j,i]=(p[j+1,i-1]-p[j,i-1])/ (x[j+i]-x[j])
  print("{0}\t{1:6.1f} \t{2:6.2f} \t {3:6.5f} \t {4:6.5f} \t {5:6.5f} \t {6:6.5f}".format(w,x[w], y[w] , p[w,1], p[w,2], p[w,3], p[w,4]))
#print(p)

#Interpolamos
xi=1.5
xp=1
yi=p[0,0]

for k in range (0,n-1):
  xp=xp*(xi-x[k])
  yi=yi+p[0,k+1]*xp #polinomio
print("")
print("El valor aproximado de",xi," es:" ,yi,)
#print(yi)

#GRAFICA

plt.plot(x,y,'o')
plt.plot(xi,yi,'o')
plt.text(xi+0.1,yi, str(yi))
plt.title('Metodo de diferencias de Newton')
plt.legend(['Puntos dados','Polinomio'])
plt.grid(True)
plt.show()
```
